```python
import pandas as pd
from plotnine import *
from PIL import Image, ImageDraw
import io
```

```python
"""
Read in the data and clean it
(data comes from https://www.smhi.se/data/meteorologi/temperatur/uppsalas-temperaturserie-1.2855)
"""

temperature = pd.read_csv('uppsala_tm_1722-2019.dat', sep=' +',engine='python',
                                 index_col=False, header=None,
                                 names=["year", "month", "day", "obs", "obscorr","loc"])


temperature['month'] = temperature['month'].astype('category')
temperature
```

```python
"""
Pick out the mean temperature over each month
"""
mean_month_temp = temperature.groupby(['year', 'month'], as_index=False)[['obs','obscorr']].mean()
mean_month_temp.loc[mean_month_temp['year']==2019]
```

```python
"""
Create gif from each years monthly mean temperature.
"""
theme_set(theme_bw())

theme_update(    
    axis_text = element_text(face="bold", size=15, angle=45, margin={"t" : 0, "r" : 0, "b" : 0, "l" : 0}),
    axis_text_x  = element_text(face = "bold", size = 15, angle = 45, margin = {"t" : 0, "r" : 0, "b" : 0, "l" : 0}),
    axis_text_y  = element_text(face = "bold", size = 15, angle = 45),
    
    plot_title   = element_text(size = 15, face = "bold.italic"),
    axis_title_x = element_text(size = 15, face = "bold", margin = {"t" : -2.5, "r" : 0, "b" : 0, "l" : 0}),
    axis_title_y = element_text(size = 15, face = "bold", margin = {"t" : 0, "r" : 0, "b" : 0, "l" : 0}),
    
    strip_text_x = element_text(size = 15),
    strip_text_y = element_text(size = 15),
    legend_title = element_text(size = 17), 
    legend_text  = element_text(size = 17)
)

min_temp = mean_month_temp['obs'].min()
max_temp = mean_month_temp['obs'].max()

images = []

def fig_to_img(plotnine_fig):
    buf = io.BytesIO()
    im = plotnine_fig.save(buf, format='png', verbose=False)
    buf.seek(0)
    im = Image.open(buf)
    im.load()
    buf.close()
    return im

for y in range(mean_month_temp['year'].min(),mean_month_temp['year'].max()):
    fig_means = (ggplot(mean_month_temp.loc[mean_month_temp['year']==y],
        aes(x='month', y='obs', group='year')) + 
        geom_line(alpha = 0.5, size = 1.5) +
        ylim(min_temp, max_temp) +
        labs(title="Year: " + str(y), x='Month', y='Average temperature') # customizing labels
    )
    
    images.append(fig_to_img(fig_means))

images[0].save('pillow_imagedraw.gif',save_all=True, append_images=images[1:], 
               optimize=False, duration=150, loop=0)
```
